```python
import pandas as pd
import matplotlib.pyplot as plt

data_dict = {
    "Sales": [10, 20, 30, 40],
    "Profits": [2, 4, 3, 1],
    "Date": ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01']
}

df = pd.DataFrame(
    data_dict,
    columns=['Sales', 'Profits'],
    index=data_dict['Date']
)
fig, ax = plt.subplots()
# the index is used as the x axis and each column is as a separate series by default
ax.plot(df)
ax.set_title('Product Sales and Profits')

# explicitly setting the x and y axis - gives more control and is preferrable
ax.plot(df.index, df['Sales'])
ax.plot(df.index, df['Profits'])
ax.set_title('Product Sales and Profits')
```
